```python
# convert the input into a array expected for LR
experience = np.array(df['exp']).reshape(-1, 1)
salary = np.array(df['salary'])
```

```python
# Train model
model = LinearRegression()
model.fit(experience, salary) #fit the independant to dependant

# Predict salary (get regression line)
predicted_salary = model.predict(experience)
print()
```

```python
# Plot
plt.scatter(experience, salary, color='blue', label='Actual data')
plt.plot(experience, predicted_salary, color='red', label='Regression line')
plt.xlabel('Experience (months)')
plt.ylabel('Salary (thousands)')
plt.title('Linear Regression: Experience vs Salary')
plt.legend()
plt.grid(True)
plt.show()
```
